3. En un solo gráfico muestre la distribución por tipo de pago (el gráfico circular que hizo anteriormente) y la serie de tiempo del total dinero por tipo de pago. 
   
Para la serie de tiempo realice los siguientes pasos:
   - Construya un dataframe con pagos únicos, es decir, que no se repitan los tipos de pago por orden.
   - Junte las ventas con las entregas para tener las fechas.
   - Construya un DataFrame que junte el id de la orden, la fecha y los tipos de pago. 
   - Construya la Fecha.
   - Construya la serie de tiempo para cada medio de pago. Puede usar `groupby` luego `pivot_table` o directamente `pivot_table`

```python
##### -- Construya un dataframe con pagos únicos, es decir, que no se repitan los tipos de pago por orden. ----- #
```

```python
pagos.head()
```

```python
pagos.groupby(['order_id','payment_type'])[['payment_value']].sum().head()
```

```python
pagos_unicos = pagos.groupby(['order_id','payment_type'])[['payment_value']].sum().reset_index()
pagos_unicos.head()
```

```python
##    - Junte las ventas con las entregas para tener las fechas.
```

```python
ventas = ventas.merge(entregas)
```

```python
ventas.head()
```

```python
#    - Construya un DataFrame que junte el id de la orden, la fecha y los tipos de pago. 

pagos_fecha = ventas.loc[:,['order_id','order_purchase_timestamp']].merge(pagos_unicos)
```

```python
pagos_fecha.head()
```

```python
### ----- Transformar a fechas --- ####
pagos_fecha['order_purchase_timestamp'] = pd.to_datetime(pagos_fecha.order_purchase_timestamp)
```

```python
### ------ Extraer solo la fecha ---- ####
pagos_fecha['order_purchase_timestamp']  = pagos_fecha.order_purchase_timestamp.dt.date
```

```python
#### ---- Crear una tabla pivote ---- ####
pagos_fecha=pagos_fecha.pivot_table(index='order_purchase_timestamp',
                                    columns='payment_type', 
                                    values='payment_value', 
                                    aggfunc=sum)
```

```python
pagos_fecha.head()
```

```python
pagos_fecha.plot(figsize=(16,8))
```

**combinar los graficos**

```python
tipos_pago
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,8))
plt.suptitle('Comportamiento de los Tipos de pago', fontsize=30)
## ----- Derecha ---- ####
colores = plt.get_cmap('rainbow')
ax2.pie(x = tipos_pago, labels= tipos_pago.index,
       textprops = {'fontsize':14},explode = (0.3, 0,0,0,0.1), 
                       shadow=False,autopct = '%.00f',
                       colors= colores(np.linspace(0,1,tipos_pago.size)))

ax2.set_title('Distribución del medio de pago', fontsize=20)

##### ----- Grafico 2: Izquierda ---- ####
pagos_fecha.plot(ax = ax1)
ax1.set_title('Serie de tiempo', fontsize=20)

plt.show()
```
